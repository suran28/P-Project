```python
import json

def load_jsonl(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            try:
                # 각 라인을 JSON 객체로 파싱하여 리스트에 추가
                json_obj = json.loads(line)
                data.append(json_obj)
            except json.JSONDecodeError as e:
                print(f"Error decoding JSON in line: {line}")
                print(e)
    return data

# 파일 경로를 적절히 변경하세요.
file_path = '/Users/suran/Desktop/gpt/validation_dataset.jsonl'
jsonl_data = load_jsonl(file_path)

# 결과 확인
for item in jsonl_data:
    print(item)
```

```python
def count_jsonl_lines(file_path):
    line_count = 0
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            line_count += 1
    return line_count

# 파일 경로를 적절히 변경하세요.
file_path = '/Users/suran/Desktop/gpt/validation_dataset.jsonl'
num_lines = count_jsonl_lines(file_path)

print(f"The number of lines in the JSONL file is: {num_lines}")
```

```python
from pathlib import Path
from openai import OpenAI
import os


os.environ['OPENAI_API_KEY']=""

client = OpenAI(
    # This is the default and can be omitted
    api_key=os.environ.get(""),
)

training = client.files.create(
    file=Path("cert_total1215.jsonl"),
    purpose="fine-tune",
)
validation = client.files.create(
    file=Path("validation_dataset"),
    purpose="fine-tune",
)

print(training.id)
print(validation.id)
```

```python
response = client.fine_tuning.jobs.create(
    training_file=training.id,
    validation_file=validation.id,
    model="gpt-3.5-turbo",
)

job_id = response.id

# You can use the job ID to monitor the status of the fine-tuning job.
# The fine-tuning job will take some time to start and complete.

print("Job ID:", response.id)
print("Status:", response.id)
print(response.model_dump_json(indent=2))
```

```python
# Track training status

from IPython.display import clear_output
import time

start_time = time.time()

# Get the status of our fine-tuning job.
response = client.fine_tuning.jobs.retrieve(job_id)

status = response.status

# If the job isn't done yet, poll it every 10 seconds.
while status not in ["succeeded", "failed"]:
    time.sleep(10)

    response = client.fine_tuning.jobs.retrieve(job_id)
    print(response.model_dump_json(indent=2))
    print("Elapsed time: {} minutes {} seconds".format(int((time.time() - start_time) // 60), int((time.time() - start_time) % 60)))
    status = response.status
    print(f'Status: {status}')
    clear_output(wait=True)

print(f'Fine-tuning job {job_id} finished with status: {status}')

# List all fine-tuning jobs for this resource.
print('Checking other fine-tune jobs for this resource.')
response = client.fine_tuning.jobs.list()
print(f'Found {len(response.data)} fine-tune jobs.')
```
