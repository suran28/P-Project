```python
import requests
import pprint

#인증키 입력
encoding = 'qUjfGRDJasifRNhDiZxLmTm%2FBUOKOgYdzyBk1RRcOg9EGlFjE3KfP2ABf6exHYkmgLoqi6IsDSsOfXuPd0bWsg%3D%3D'
decoding = 'qUjfGRDJasifRNhDiZxLmTm/BUOKOgYdzyBk1RRcOg9EGlFjE3KfP2ABf6exHYkmgLoqi6IsDSsOfXuPd0bWsg=='

url = 'http://openapi.q-net.or.kr/api/service/rest/InquiryMjrQualSVC/getList?'

acqu_2022 = pd.DataFrame()
grdCd_list = ['10','20','30','31','32','33','39','40','41']

for i in grdCd_list :
    params ={'serviceKey' : decoding,
             'grdCd': i,
             'baseYY':'2022',
            'pageNo':'1',
            'numOfRows':'300'}

    response = requests.get(url, params=params)

    contents = response.text

    pp = pprint.PrettyPrinter(indent = 4)
    xml_obj = bs4.BeautifulSoup(contents,'lxml-xml')
    rows = xml_obj.findAll('item')
    row_list = [] # 행값
    name_list = [] # 열이름값
    value_list = [] #데이터값

    # xml 안의 데이터 수집
    for i in range(0, len(rows)):
        columns = rows[i].find_all()
        #첫째 행 데이터 수집
        for j in range(0,len(columns)):
            if i ==0:
                # 컬럼 이름 값 저장
                name_list.append(columns[j].name)
            # 컬럼의 각 데이터 값 저장
            value_list.append(columns[j].text)
        # 각 행의 value값 전체 저장
        row_list.append(value_list)
        # 데이터 리스트 값 초기화
        value_list=[]
        acqu_df = pd.DataFrame(row_list, columns=name_list)
        
        #데이터프레임 합치기
        acqu_2022 = pd.concat([acqu_2022, acqu_df])
acqu_2022
```
